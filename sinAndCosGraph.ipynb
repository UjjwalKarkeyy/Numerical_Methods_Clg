```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import cmath
from sympy import symbols, Eq, solve
```

```python
trigonometric_functions = ["sin", "cos"]

for func in trigonometric_functions:
    x = np.arange(0.0,5*np.pi,0.1)
    if func == "sin":
        y = 2 * np.sin(x)
    elif func == "cos":
        y = np.cos(x)
    fig = plt.figure(figsize=(15,7))
    plt.plot(x,y, marker='o',color = 'blue')
    plt.grid()
    plt.show()
```

```python
# Finding quadratic roots

equation = "2xsq + 5x - 25"

# Split the equation into coefficients
coefficients = equation.split("x")
a = int(coefficients[0])
b= int(coefficients[1].split("+")[1])
c = int(coefficients[2].split('-')[1])
# print(a,b,c)


def cal_root(a, b, c):
    discriminant = b**2 - 4*a*c
    sqrt_d = cmath.sqrt(discriminant)  # use cmath to support complex numbers
    root1 = (-b + sqrt_d) / (2 * a)
    root2 = (-b - sqrt_d) / (2 * a)
    return root1, root2

answer_list = cal_root(a,b,c)
print(answer_list)
```

```python
# Find equation of line passing through A(1,2), B(5,9)
A = (1, 2)
B = (5, 9)
m = (B[1] - A[1]) / (B[0] - A[0])
c = B[1] - m * B[0]
print(f"Equation of line is y = {m}x + {c}")
```

```python
# Solve equations:
# 2x + 3y = 7
# 3x - 6y = 10

x, y = symbols('x,y')

eq1 = Eq((2*x+3*y),7)
print(f"Equation 1:{eq1}")
eq2 = Eq((3*x-6*y),10)
print(f"Equation 2:{eq2}")
solution = solve((eq1,eq2), (x,y))
```
