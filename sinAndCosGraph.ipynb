```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.arange(0.0,5*np.pi,0.1)
y = 2*(np.sin(x))
fig = plt.figure(figsize=(15,7))
plt.plot(x,y, marker='o',color = 'blue')
plt.grid()
plt.show()
```

```python
x = np.arange(0.0,5*np.pi,0.1)
y = np.cos(x)
fig = plt.figure(figsize=(11,7))
plt.plot(x,y, marker='o',color = 'blue')
plt.grid()
plt.show()
```
